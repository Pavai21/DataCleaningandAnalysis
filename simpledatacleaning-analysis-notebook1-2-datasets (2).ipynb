**Finding out the duplicates in dataset**

```python
df_data.duplicated().sum()
```

```python
duplicate = df_data[df_data.duplicated()]
```

```python
duplicate
```

**Dropping the duplicate values**

```python
drop_duplicate = df_data.drop_duplicates()
```

```python
drop_duplicate
```

**Filtering the necessary columns for analyse**

```python
Filtered_Data = drop_duplicate.filter(['country','director_name','title_year','genres','budget','imdb_score'])
```

```python
Filtered_Data
```

```python
df2 = Filtered_Data.sort_values(by=['title_year','budget','imdb_score'],ascending = False)
```

```python
df2
```

```python
df2['genres'].value_counts()
```

**Getting the budget value greater than 200000000**

```python
df3 = df2[(df2['budget']>200000000)]
```

```python
df4 = df3.head(20)
```

```python
df4
```

**Visualising Country vs Budget using matplotlib**

```python
import matplotlib.pyplot as plt
```

```python
x = df4['country']
y = df4['budget']
```

```python
plt.figure(figsize=(10,9))
plt.bar(x,y)
plt.title("Country vs Budget",fontsize =20)
plt.xlabel("Country", fontsize = 12)
plt.ylabel("Budget", fontsize = 12)
```
